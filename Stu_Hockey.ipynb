```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.craigslist_db
collection = db.hockey
```

```python
# URL of page to be scraped
url = 'https://www.nhl.com/'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('div', class_="mixed-feed__item-header-text")
```

```python
# Loop through results to retrieve article title, header, and timestamp of article

for result in results:
    data = {}
    title = result.find('h4', class_='mixed-feed__header').text
    lede = result.find('h5', class_='mixed-feed__subheader').text
    print(title)
    print(lede)
    # Dictionary to be inserted into MongoDB
    data[" title"] = title
    data[" lede"] = lede
    collection.insert_one(data)

    # Insert dictionary into MongoDB as a document
    
```

```python
# Display the MongoDB records created above
articles = db.hockey.find()
for article in articles:
    print(article)
    print("\n")
```
